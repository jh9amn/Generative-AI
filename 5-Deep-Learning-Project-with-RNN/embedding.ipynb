```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'I love deep learning',
    'RNNs are great for sequence data',
    'Embeddings help represent words as vectors',
    'I enjoy coding in Python',
    'I am learning about RNNs and embeddings',
    'You can build powerful models with deep learning',
    'We are using TensorFlow and Keras',
    'I like to experiment with neural networks'
]
```

```python
sent
```

```python
## Define a vocabulary size
vocab_size = 10000
```

```python
## One Hot Representation
one_hot_repr = [one_hot(words, vocab_size) for words in sent]
one_hot_repr
```

```python
## Word Embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import pad_sequences
```

```python
import numpy as np
```

```python
sent_length = 8
embedding_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
print(embedding_docs)
```

```python
## Feature Representation Dimension
dim = 10
```

```python
# model = Sequential()
# model.add(Embedding(vocab_size, dim, input_length=sent_length))
# model.compile(optimizer='adam', loss='mse')


model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=dim, input_length=sent_length))
# model.add(SimpleRNN(32))
# model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Build model (optional, but ensures summary works)
model.build(input_shape=(None, sent_length))
```

```python
model.summary()
```
