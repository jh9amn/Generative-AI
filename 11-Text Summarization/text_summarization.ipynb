```python
chat_messages = [
    SystemMessage(content="You are a helpful assistant that summarizes speeches."),
    HumanMessage(content=f'Please provide a short and concise summary of the following speech:\n Text: """{speech}"""'),
]
```

```python
llm = ChatGroq(groq_api_key=groq_api_key, model="llama-3.1-8b-instant")
```

```python
llm.get_num_tokens(speech)
```

```python
response = llm.invoke(chat_messages)

# Access the content
print(response.content)
```

#### Prompt Template Text Summarization

```python
from langchain_classic.chains import LLMChain
from langchain_core.prompts import PromptTemplate

generictemplate = """
Write a summary of the following speech:
Speech: {speech}
Translate the precise summary to {language}.
"""

prompt = PromptTemplate(
    input_variables=["speech", "language"],
    template=generictemplate,
)

prompt
```

```python
complete_prompt = prompt.format(speech=speech, language="hindi")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
summary = llm_chain.run({"speech" : speech, "language": "hindi"})
summary
```

## Stuff Document Chain Text Summanrization

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_core.prompts import PromptTemplate
loader = PyPDFLoader("apjspeech.pdf")
pages = loader.load_and_split()
pages
```

```python
## prompt for summarization of pdf
template = """ 
Write a concise and short summary of the following speech,
Speech: {text}
"""

prompt = PromptTemplate(input_variables=["text"], template=template)
```

```python
from langchain_classic.chains import load_summarize_chain
```

```python
chain = load_summarize_chain(llm, chain_type="stuff", prompt=prompt, verbose=True)
output_summary = chain.run(pages)
output_summary
```

## Map Reduce to Summarize large documents

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
final_docs = RecursiveCharacterTextSplitter(
    chunk_size = 2000,
    chunk_overlap  = 100
).split_documents(pages)

final_docs
```

```python
len(final_docs)
```

```python
prompt
```

```python
chunks_prompt = """
Please summarize the below speech:
Speech: {text}
Summary:
"""

map_prompt_template = PromptTemplate(input_variables=["text"], template=chunks_prompt)
```

```python
final_prompt = """
Provide the final summary of the entire speech with these important points.
Add a motivation Title, Start the precise summary with an introduction and provide the summmary in number 
points for the speech.capitalize
Speech: {text}
"""

final_prompt_template = PromptTemplate(
    input_variables=["text"],
    template=final_prompt
)

final_prompt_template
```

```python
summary_chain = load_summarize_chain(
    llm, 
    chain_type="map_reduce",
    map_prompt = map_prompt_template,       ## first prompt template give summary of each chunk
    combine_prompt = final_prompt_template,  ## second prompt template to combine the summary of each chunk and give the final summary
    verbose=True
)

output_summary = summary_chain.run(final_docs)
output_summary
```

## Refine Chain For Summarization

```python
chain = load_summarize_chain(
    llm,
    chain_type="refine",
    verbose=True
)


output_summary = chain.run(pages)
output_summary
```
