```python
from langchain_groq import ChatGroq
import os
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
from langchain_core.messages import (
    AIMessage,
    HumanMessage,
    SystemMessage,
)
```

```python
speech = """My fellow citizens and young innovators of this great nation, Today, I stand before you not just as your representative, but as a believer in the limitless potential of the Indian spirit. We are a nation that is no longer waiting for the future; we are the ones writing it. As someone rooted in the world of technology, I see a Bharat where every village is connected by high-speed fiber, and every citizen is empowered by digital literacy. Our goal is to move from being the back-office of the world to becoming the Innovation Hub of the planet. We are transitioning from 'Code' to 'Product,' and from 'Service' to 'Creation'. True independence comes from self-reliance. Whether it is our defense systems, our semiconductors, or our software—'Made in India' must become the global gold standard for quality and trust. We are investing in our youth, our startups, and our scientists because the next Google, the next SpaceX, or the next global AI giant must be born on Indian soil. Technology is the greatest equalizer. Through digital governance, we are ensuring that every rupee meant for the poor reaches their bank accounts directly, without leakage. Transparency is our tool, and 'Minimum Government, Maximum Governance' is our mantra. Let us pledge to build a Viksit Bharat—a developed India—where tradition meets technology and where every Indian has the wings to fly. Together, we shall lead the world in this new era. Jai Hind!"""
```

```python
speech
```

```python
chat_messages = [
    SystemMessage(content="You are a helpful assistant that summarizes speeches."),
    HumanMessage(content=f'Please provide a short and concise summary of the following speech:\n Text: """{speech}"""'),
]
```

```python
llm = ChatGroq(groq_api_key=groq_api_key, model="llama-3.1-8b-instant")
```

```python
llm.get_num_tokens(speech)
```

```python
response = llm.invoke(chat_messages)

# Access the content
print(response.content)
```

#### Prompt Template Text Summarization

```python
from langchain_classic.chains import LLMChain
from langchain_core.prompts import PromptTemplate

generictemplate = """
Write a summary of the following speech:
Speech: {speech}
Translate the precise summary to {language}.
"""

prompt = PromptTemplate(
    input_variables=["speech", "language"],
    template=generictemplate,
)

prompt
```

```python
complete_prompt = prompt.format(speech=speech, language="hindi")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
summary = llm_chain.run({"speech" : speech, "language": "hindi"})
summary
```
